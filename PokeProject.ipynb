```python
poke_data = []
for i in range(1,152):
    poke_data.append([
        pb.pokemon(i).name,
        pb.pokemon(i).id,
        pb.pokemon(i).height,
        pb.pokemon(i).weight
    ])

poke_df = pd.DataFrame(poke_data, columns = ['name', 'id', 'height', 'weight'])
poke_df.head()
```
